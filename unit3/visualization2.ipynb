```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

people = pd.read_csv('people.csv')
people
```

```python
plt.figure(figsize=(16,9))

sns.regplot(data=people, x='gpa', y='friends')
```

```python
sns.lmplot(data = people, x = 'gpa', y = 'friends', hue = 'footware')
```

```python
wine_reviews = pd.read_csv('winemag-data-130k-v2.csv')
wine_reviews
```

```python
best_wines = wine_reviews.loc[wine_reviews.points == 100].value_counts('country')

sns.barplot(data=wine_reviews, x= best_wines.index, y=best_wines)
```
