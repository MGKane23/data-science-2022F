```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

sea_levels = pd.read_csv('epa-sea-level.csv')
sea_levels.rename({'CSIRO Adjusted Sea Level': 'sea-level'}, axis='columns', inplace=True)
sea_levels.info()
```

```python
sns.regplot(data=sea_levels, x = 'Year', y = 'sea-level')
```

```python
X = sea_levels.Year.values.reshape(-1,1)
y = sea_levels['sea-level'].values.reshape(-1,1)

model = LinearRegression().fit(X, y)

print('slope:', model.coef_)
print('y-intercept:', model.intercept_)
print('coefficient of determination (r squared value):', model.score(X,y))
```

```python
predict_2050 = model.predict([[2050]])
predict_2070 = model.predict([[2070]])
predict_2100 = model.predict([[2100]])

print('The sea level in the year 2050 is predicted to be:', predict_2050[0][0], 'inches.')
print('The sea level in the year 2070 is predicted to be:', predict_2070[0][0], 'inches.')
print('The sea level in the year 2100 is predicted to be:', predict_2100[0][0], 'inches.')
```

```python
level_2013 = sea_levels.loc[sea_levels.Year == sea_levels.Year.max(), 'sea-level'].values[0]

rise_by_2050 = predict_2050[0][0] - level_2013 
rise_by_2070 = predict_2070[0][0] - level_2013 
rise_by_2100 = predict_2100[0][0] - level_2013 

print('In year 2050 the sea level will be', rise_by_2050, 'inches higher than in 2013.')
print('In year 2070 the sea level will be', rise_by_2070, 'inches higher than in 2013.')
print('In year 2100 the sea level will be', rise_by_2100, 'inches higher than in 2013.')
```
